```python
# Importing Library

import pandas as pd
```

```python
# Importing data from MS Excel File

assignments_df = pd.read_excel("data.xlsx", sheet_name="Assignments")
project_df = pd.read_excel("data.xlsx", sheet_name="Projects")
attendance_df = pd.read_excel("data.xlsx", sheet_name="Attendance")
grades_df = pd.read_excel("data.xlsx", sheet_name="Grades")
```

```python
# Code to find Project Score

final_project_score = 0
for i in project_df.index:
    project_score = (project_df["Score"][i]*(project_df["Percentage"][i]))/100
    final_project_score = final_project_score + project_score
final_project_score
```

```python
# Code to find Attendance Score

final_attendance_score = attendance_df[attendance_df["Attendance"] == "P"]["Class"].count()*(10/15)
final_attendance_score
```

```python
# Code to find Assignment Score

random_assignment_df = assignments_df.sample(n=10)
final_assignment_score = random_assignment_df["Score"].sum()*0.03
final_assignment_score
```

```python
# Code to find Final Score

final_score = final_attendance_score + final_assignment_score + final_project_score
final_score
```

```python
grades_df
```

```python
# Code to find the Grade based on Score

for i in grades_df.index:
    if grades_df["Min"][i]<=final_score<=grades_df["Max"][i]:
        print(grades_df["Grade"][i])
```
