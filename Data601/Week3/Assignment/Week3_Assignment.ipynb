```python
# Importing Libraries
import pandas as pd
import numpy as np
import faker
```

```python
# Initializing faker
faker = faker.Faker()
```

```python
# Function to create fake profiles
def make_fake_profiles(num):
    
    owns_or_rents = ["Owns", "Rents"]
    
    fake_profiles = [{
        'Email': faker.email(),
        'Phone Numbers': faker.phone_number(),
        'Home Address': faker.address(),
        'Name': faker.name(),
        'Date of Birth': faker.date_between(start_date='-60y', end_date='today'),
        'Number of Kids': np.random.randint(1,5),
        'Owns or Rents': np.random.choice(owns_or_rents),
        'Annual Income':np.random.randint(5000,1000000),
        'Speeding Tickets in past year':np.random.randint(0,20), 
    } for x in range(num)]
    
    return fake_profiles
```

```python
# Input & Result
profiles_df = pd.DataFrame(make_fake_profiles(num=int(input("Enter number of profiles you want to create: "))))
profiles_df.to_csv("fake_data.csv")
profiles_df
```
