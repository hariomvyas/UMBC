# data 1
https://data.ny.gov/Energy-Environment/Solid-Waste-Management-Facilities/2fni-raj8

```python
df1 = pandas.read_csv('https://data.ny.gov/api/views/2fni-raj8/rows.csv?accessType=DOWNLOAD&sorting=true')
df1.shape
```

```python
df1.head(5)
```

### Is this data tidy or not?

(Yes or no)

*if the data is not tidy, then convert the data to a tidy dataframe*

##### Answer -
    No

```python
id_vars = [
    'Facility Name', 'Location Address', 'Location Address2', 'City',
       'State', 'Zip Code', 'County', 'Region', 'Phone Number', 'Owner Name',
       'Owner Type', 'Activity Desc', 'Activity Number', 'Active',
       'East Coordinate', 'North Coordinate', 'Accuracy Code', 'Waste Types',
       'Authorization Number', 'Authorization Issue Date', 'Expiration Date',
       'Georeference'
]
```

```python
## Learned from here - https://www.pauldesalvo.com/how-to-break-up-a-comma-separated-string-in-a-pandas-column/

## Split Waste Types
waste_types_expanded = df1['Waste Types'].str.split(';', expand=True)
waste_types_expanded.columns = ['W'+str(i) for i in waste_types_expanded.columns]

## Concating old Dataframe & new waste types dataframe
concated_df = pandas.concat([df1, waste_types_expanded], axis = 1)

final_df1 = pandas.melt(concated_df, id_vars, value_vars=waste_types_expanded.columns, var_name='Country Number', value_name='Waste Types')

## Dropping Extra Columns
final_df1.drop(final_df1.columns[[22]], axis = 1, inplace=True)
```

```python
final_df1
```

```python
print(final_df1.shape)
print(final_df1.columns)
```
