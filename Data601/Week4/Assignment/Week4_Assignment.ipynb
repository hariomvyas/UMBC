```python
# Importing Libraries

from bs4 import BeautifulSoup
from lxml import etree
```

#### Previous Code had some varibale problems which i fixed. 

```python
# Parsing XML & HTML File

tree = etree.parse("xml_containing_html.xml")
root = tree.getroot()
xml_string = etree.tostring(root)
soup = BeautifulSoup(xml_string, 'html.parser')
all_pages = soup.find('all_pages')
```

```python
# Finding count of links and total count

total_count = 0
print("Count of Links per HTML File is: ")
for page in all_pages:
    print(page.text.count("</a>"))
    total_count = total_count + int(page.text.count("</a>"))
print("Total count of Link is: ",total_count)
```
